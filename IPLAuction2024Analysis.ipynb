```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
path="/content/drive/MyDrive/Auction2024.csv"
df=pd.read_csv(path)
```

```python
#Total Number OF Players In Each Category
df['TYPE'].value_counts()
```

```python
#Total Number OF Indian And Overseas Player
df['NATIONALITY'].value_counts()
```

```python
#Highest Paid Player in Auction
print("Highest Paid Player in Auction:")
print(df[df['PRICE PAID'] == df['PRICE PAID'].max()])
```

```python
#Lowest Paid Player In Auction
print("Lowest Paid Player In Auction:")
print(df[df['PRICE PAID'] == df['PRICE PAID'].min()])
```

```python
#Most Number of Players Picked by Team
print("Most Number of Overseas Picked by Team:")
print(df[df['TEAM'] == df['TEAM'].value_counts().idxmax()])
```

```python
#Lowest Number of Players Picked By a Team
print("Lowest Number of Overseas Picked By a Team:")
print(df[df['TEAM'] == df['TEAM'].value_counts().idxmin()])
```

```python
#Total Money Spent By Team in Auction
total_money_spent = df.groupby('TEAM')['PRICE PAID'].sum().sort_values(ascending=False)
print("Total Money Spent By Team in Auction:")
print(total_money_spent)
```

```python
#Total Money Spent By Team in Auction
total_money_spent = df.groupby('TEAM')['PRICE PAID'].sum().sort_values(ascending=False)
print("Total Money Spent By Team in Auction:")
print(total_money_spent)
```

#**@GRAPHS**


```python
#Distribution of Nationalities
plt.figure(figsize=(8, 3))
sns.countplot(x='NATIONALITY', data=df, order=df['NATIONALITY'].value_counts().index)
plt.title('Distribution of Nationalities')
plt.xlabel('Nationality')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
#Distribution OF Price Paid
plt.hist(df['PRICE PAID'], bins=20)
plt.xlabel('Price Paid')
plt.ylabel('Frequency')
plt.title('Distribution of Price Paid')
plt.show()
```

```python
#Distribution Of Teams
plt.figure(figsize=(20, 6))
sns.countplot(x='TEAM', data=df, order=df['TEAM'].value_counts().index)
plt.title('Distribution of Teams')
plt.xlabel('Team')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
